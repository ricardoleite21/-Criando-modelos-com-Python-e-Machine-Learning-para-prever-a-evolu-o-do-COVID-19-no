```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
import pandas as pd

# Abre o widget para upload de arquivos
uploaded = files.upload()

# Verifica o nome do arquivo e carrega-o em um DataFrame do pandas
for filename in uploaded.keys():
    print('Uploaded file name:', filename)
    df = pd.read_excel(filename)

# Verifica as primeiras linhas do DataFrame
print(df.head())
```

```python
# Mostre as informações gerais sobre o conjunto de dados
print(df.info())

# Mostre estatísticas descritivas
print(df.describe())
```

```python
# Verifique valores ausentes
print(df.isnull().sum())

# Verifique duplicatas
print(df.duplicated().sum())
```

```python
# Variáveis categóricas
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(df[col].value_counts())
    print("\n")
```

```python
# Variáveis numéricas
numeric_columns = df.select_dtypes(include=[np.number]).columns
for col in numeric_columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribuição de {col}')
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Criação de um DataFrame de exemplo
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [9, 8, 7, 6, 5]
}

df = pd.DataFrame(data)

# Exibir o DataFrame
print("DataFrame de exemplo:")
print(df)

# Seleciona apenas as colunas numéricas (neste exemplo, todas são numéricas)
df_numeric = df.select_dtypes(include=[float, int])

# Calcula a matriz de correlação
correlation_matrix = df_numeric.corr()

# Exibir a matriz de correlação
print("\nMatriz de Correlação:")
print(correlation_matrix)

# Plota a matriz de correlação
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```
